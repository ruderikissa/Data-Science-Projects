```python
def PlotPolly(model,independent_variable, dependent_variable, Name, x1, x2):
    
    x_new = np.linspace(x1,x2,100)
    y_new = model(x_new)
    
    plt.plot(independent_variable, dependent_variable, ".", x_new, y_new ,"-")
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.title("Plynomial Fit")
    plt.xlabel(Name)
    plt.ylabel(price)
    
    
    plt.show()
    plt.close()
```

```python
x = df1[CW]
y = df1[price]
```

```python
# Multiple Linear Regression (Hp,CW,Br,L,W)

X1 = df1[[Hp,CW,Br,L,W]]
lm2 = LinearRegression()
lm2.fit(X1,df1[price])
Y_hat = lm2.predict(X1)
mse_M = mean_squared_error(Y,Y_hat)
print("The R-square is: ", lm2.score(X1,Y))
print("The mean squared error is: ", mse_M)
```

```python
# Multiple Lineal Regression (ES,WB,Br)

X2 = df1[[ES,WB,Br]]
lmM = LinearRegression()
lmM.fit(X2,Y)
Y_hat1 = lmM.predict(X2)
mse_M2 = mean_squared_error(Y,Y_hat1)
print("The R-square is: ", lmM.score(X2,Y))
print("The mean squared error is: ", mse_M2)
```
