```python
df[["normalized-losses"]] = df[["normalized-losses"]].astype("float")
df[["bore","stroke"]] = df[["bore","stroke"]].astype("float")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
df[["horsepower"]] = df[["horsepower"]].astype("float")
```

```python
df[df.isnull().any(axis = 1)]
```

```python
nl_mean = df["normalized-losses"].mean() 
df["normalized-losses"].replace(np.nan, nl_mean, inplace = True)
```

```python
df[df.notnull().any(axis = 1)].count()
```

```python
df["normalized-losses"].isnull().value_counts()
```

```python
bore_avg = df["bore"].mean()
df["bore"] = df["bore"].replace(np.nan, bore_avg)
```

```python
print(df["num-of-doors"].value_counts().idxmax())
df[df["num-of-doors"].isna()]
```

```python
df["num-of-doors"] = df["num-of-doors"].replace(np.nan, "four")
```

```python
avg_horsepower = int(df["horsepower"].mean())
avg_peak_rpm = int(df["peak-rpm"].mean())
avg_stroke = df["stroke"].mean()
```

```python
df["horsepower"].replace(np.nan,avg_horsepower, inplace = True)
df["peak-rpm"].replace(np.nan,avg_peak_rpm, inplace = True)
df["stroke"].replace(np.nan,avg_stroke, inplace = True)
```

```python
df[df.isnull().any(axis = 1)]
```

```python
df.dropna(subset = ["price"], axis = 0, inplace = True)
```

```python
df[df.isnull().any(axis = 1)]
```

```python
df[["horsepower", "peak-rpm", "price", "normalized-losses"]] = df[["horsepower", "peak-rpm", "price", "normalized-losses"]].astype("int")
```

```python
df.reset_index(drop = True, inplace = True)
```

```python
df.head()
```

```python
df["city-L/100km"] = 284/df["city-mpg"]
df["highway-L/100km"] = 284/df["highway-mpg"]
df.head()
```

```python
df.rename(columns = {'city-mpg' : 'city-L/100km', 'highway-mpg': 'highway-L/100km' }, inplace = True)
df.head()
```

```python
df["height"] = df["height"]/df["height"].max()
df["length"] = df["length"]/df["length"].max()
df["width"] = df["width"]/df["width"].max()
df[["length", "width", "height"]].head()
```

```python
bins = np.linspace(min(df["horsepower"]),max(df["horsepower"]), 4)
bins
```

```python
hp_labels = ["low", "medium", "high"]
df["hp-binned"] = pd.cut(df["horsepower"], bins, labels = hp_labels)
df[["horsepower","hp-binned"]].head(10)
```

```python
import matplotlib.pyplot as plt
plt.bar(hp_labels,df["hp-binned"].value_counts())
plt.xlabel("Horsepower")
plt.ylabel("Quantity")
plt.title("Hp categories")
```
