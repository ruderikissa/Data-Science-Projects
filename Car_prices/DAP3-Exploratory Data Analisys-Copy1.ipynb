```python
df_test3 = df[["drive-wheels","price"]]
df_gp_t3 = df_test3.groupby(["drive-wheels"], as_index = False)
df_gp_t3.head(5)
```

```python
df_gp_t3.get_group("4wd")
```

```python
# ANOVA Test

f_score,p_score = stats.f_oneway(df_gp_t3.get_group("4wd")["price"],df_gp_t3.get_group("fwd")["price"],df_gp_t3.get_group("rwd")["price"])
print("F score =",f_score,"P value =",p_score)
```

```python
# Fwd vs 4wd
f_score,p_score = stats.f_oneway(df_gp_t3.get_group("4wd")["price"],df_gp_t3.get_group("fwd")["price"])
print("F score =",f_score,"P value =",p_score)
```

```python
# Fwd vs rwd

f_score,p_score = stats.f_oneway(df_gp_t3.get_group("fwd")["price"],df_gp_t3.get_group("rwd")["price"])
print("F %score =",f_score,"P value =",p_score)
```

```python
# 4wd vs rwd
f_score,p_score = stats.f_oneway(df_gp_t3.get_group("4wd")["price"],df_gp_t3.get_group("rwd")["price"])
print("F score =",f_score,"P value =",p_score)
```
